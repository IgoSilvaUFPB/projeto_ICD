# ***Which parallel would divide Brazil so that half of the population would be in the north and the other half in the south?***

## Importing the Dataset and Libraries: 游닀

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
import seaborn as sns
```

```python
df = pd.read_csv('BRAZIL_CITIES_CLEAN.csv')
df
```

## **First steps**游늼

### ***We will use the following columns to generate our conclusions***

+ CITY/STATE.
+ IBGE_RES_POP/CUMULATIVE_POP
+ LAT/LONG/AREA

```python
df_split_lat = df[["CITY","IBGE_RES_POP","LAT","LONG","AREA"]].sort_values("LAT")
df_split_lat
```

## *Initial Analysis*游닀

### Calculating Half of the Population Value

```python
total_population = df["IBGE_RES_POP"].sum()
half_population = total_population / 2
half_population
```

### Calculating the Cumulative Sum of Population from South to North

```python
df_split_lat["CUMULATIVE_POP"] = df_split_lat["IBGE_RES_POP"].cumsum()
df_split_lat
```

```python
# calculating the split latitude
split_latitude = df_split_lat.loc[df_split_lat["CUMULATIVE_POP"] >= half_population, "LAT"].iloc[0]
degrees = int(abs(split_latitude))
minutes = int((abs(split_latitude) - degrees) * 60)
seconds = (abs(split_latitude) - degrees - minutes/60) * 3600
dms = "Split Latitude: {} degrees {} minutes {:.2f} seconds South".format(degrees, minutes, seconds)
dms
```

## **Graphics analysis and conclusions**游늵九덢잺

```python
# setting palette and style
sns.set(style="whitegrid", palette="pastel", color_codes=True)
```

```python
# plotting the split latitude line
fp = "SPLIT_BRAZIL/BRA_adm1.shp"
map_df = gpd.read_file(fp)
map_df.plot()
plt.axhline(y=-degrees, color='r', linestyle='-')
plt.show()
```

**Listing the 10 Closest Cities to the split latitude**

```python
target_latitude = split_latitude
df_split_lat["DISTANCE_FROM_TARGET"] = np.abs(df_split_lat["LAT"] - target_latitude)
closest_cities = df_split_lat.nsmallest(10, "DISTANCE_FROM_TARGET").sort_values("LONG")
closest_cities
```

**How much area is dedicated to each group?**

```python
# calculating the area of the northern and southern regions
area_above_latitude_df = df_split_lat[df_split_lat["LAT"] >= split_latitude]
area_sum_above_latitude = area_above_latitude_df["AREA"].sum()
area_below_latitude_df = df_split_lat[df_split_lat["LAT"] < split_latitude]
area_sum_below_latitude = area_below_latitude_df["AREA"].sum()
total_area = df_split_lat["AREA"].sum()
sum_area = "The half that is to the north as: {:.2f}% of the total area. While the remaining half has: {:.2f}% of the area.".format((area_sum_above_latitude/total_area)*100, (area_sum_below_latitude/total_area)*100)
sum_area
```

```python
# plotting pie chart of the area distribution
labels = 'Northern Region', 'Southern Region'
sizes = [area_sum_above_latitude, area_sum_below_latitude]
explode = (0.1, 0)  # only "explode" the 1st slice (i.e. 'Northern Region')
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=140)
plt.show()
```

*It's noteworthy that even with an equal population distribution between the northern and southern regions, the significant disparity in land area between the two regions becomes apparent. Specifically, the northern region encompasses six times the land area of the southern region*
