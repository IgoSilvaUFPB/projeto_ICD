#  ***Which line of business has the most impact on a city's GDP?***


+ In this topic we will make correlations between GVA in different areas.

> GDP stands for Gross Domestic Product, and it is a widely used measure of the economic activity within a country. It represents the total value of all goods and services produced in a country over a specific period, typically a year.


>GVA, on the other hand, stands for Gross Value Added. It is a related concept to GDP and measures the value generated by individual sectors or industries of an economy. GVA represents the difference between the value of goods and services produced by a sector or industry and the cost of inputs used in the production process. It essentially shows the contribution of each sector to the overall GDP.

## Importing the Dataset and Libraries: üìñ

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('BRAZIL_CITIES_CLEAN.csv')
df
```

## **First steps**üìë

### ***We will use the following columns to generate our conclusions***

+ CITY/STATE.
+ GVA_AGROPEC,GVA_INDUSTRY,GVA_SERVICES,GVA_PUBLIC, GVA_TOTAL, GVA_MAIN
+ GDP, POP_GDP, GDP_CAPITA
+ IDHM
+ IBGE_RES_POP
+ COMP_TOT

```python
#Filtering the dataset

df = df[["CITY","STATE","GVA_AGROPEC","GVA_INDUSTRY","GVA_SERVICES","GVA_PUBLIC", "IBGE_RES_POP", "IDHM", "GVA_TOTAL", "GDP", "POP_GDP", "GDP_CAPITA", "GVA_MAIN", "COMP_TOT"]]
df
```

```python
df_GDP = df[["CITY","STATE","GVA_AGROPEC","GVA_INDUSTRY","GVA_SERVICES","GVA_PUBLIC"]]
df_GDP_state = df_GDP.groupby('STATE').sum()
df_GDP_state
```

```python
#Total companies by state.

df_comp = df[["STATE","COMP_TOT"]]
df_comp = df_comp.groupby('STATE').sum()
df_comp
```

```python
# IDHM per state

df_IDH = df[["STATE","IDHM"]]
df_IDH = df_IDH.groupby('STATE').mean()
df_IDH
```

## *Initial Analysis*üìñ

```python
#Graphic of the companies by state.

df_comp.plot(kind='bar', stacked=True, figsize=(10, 5), color='orange')
plt.title('GVA by State')
plt.ylabel('GVA (milions)')
plt.xlabel('State')
```

```python
#IDHM by state.

df_IDH.plot(kind='bar', stacked=True, figsize=(10, 5), color='blue')
plt.ylim(0.45, 1)
plt.title('IDHM by State')
plt.ylabel('IDHM')
plt.xlabel('State')
```

```python
# Stacked bar chart of GVA by state

df_GDP_state.plot(kind='bar', stacked=True, figsize=(10, 5))
plt.title('GVA by State')
plt.ylabel('GVA (milions)')
plt.xlabel('State')
```

Through this graph, it is evident that services play a significantly larger role in the GDP of most states, and this trend can be extended to cities and even the entire country.


*Now, we will make some correlations.*

```python
correc_services = correlation = df['GVA_SERVICES'].corr(df['GDP'])
correc_public = correlation = df['GVA_PUBLIC'].corr(df['GDP'])
correc_agropec = correlation = df['GVA_AGROPEC'].corr(df['GDP'])
correc_industry = correlation = df['GVA_INDUSTRY'].corr(df['GDP'])

print(f'Correlation with GDP and GVA_AGROPEC: {correc_agropec}')
print(f'Correlation with GDP and GVA_INDUSTRY: {correc_industry}')
print(f'Correlation with GDP and GVA_SERVICES: {correc_services}')
print(f'Correlation with GDP and GVA_PUBLIC: {correc_public}')
```

```python
correc_services = correlation = df['GVA_SERVICES'].corr(df['IDHM'])
correc_public = correlation = df['GVA_PUBLIC'].corr(df['IDHM'])
correc_agropec = correlation = df['GVA_AGROPEC'].corr(df['IDHM'])
correc_industry = correlation = df['GVA_INDUSTRY'].corr(df['IDHM'])

print(f'Correlation with IDHM and GVA_AGROPEC: {correc_agropec}')
print(f'Correlation with IDHM and GVA_INDUSTRY: {correc_industry}')
print(f'Correlation with IDHM and GVA_SERVICES: {correc_services}')
print(f'Correlation with IDHM and GVA_PUBLIC: {correc_public}')
```

```python
correc_services = correlation = df['GVA_SERVICES'].corr(df['COMP_TOT'])
correc_public = correlation = df['GVA_PUBLIC'].corr(df['COMP_TOT'])
correc_agropec = correlation = df['GVA_AGROPEC'].corr(df['COMP_TOT'])
correc_industry = correlation = df['GVA_INDUSTRY'].corr(df['COMP_TOT'])

print(f'Correlation with COMP_TOT and GVA_AGROPEC: {correc_agropec}')
print(f'Correlation with COMP_TOT and GVA_INDUSTRY: {correc_industry}')
print(f'Correlation with COMP_TOT and GVA_SERVICES: {correc_services}')
print(f'Correlation with COMP_TOT and GVA_PUBLIC: {correc_public}')
```

**Through the correlations carried out, it is possible to see that the GDP and the total number of companies have a very strong correlation with some GVA columns.**

**The MHDI column has a low correlation, therefore, it is impossible to generate any really important analysis.**
