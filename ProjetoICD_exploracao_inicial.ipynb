# **Importing the Dataset and opening remarks**

## Importing the Dataset and Libraries: üìñ

```python
import pandas as pd
import seaborn as sns   
import matplotlib.pyplot as plt
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/IgoSilvaUFPB/projeto_ICD/main/BRAZIL_CITIES.csv'
df = pd.read_csv(url, sep = ';')
df
```

## Opening Remarks: üß†

```python
print('N√∫mero de linhas do dataset: ', df.shape[0])
print('N√∫mero de colunas do dataset: ', df.shape[1])
```

```python
colunas_numericas = df.select_dtypes(include = ['float64', 'int64']).columns.tolist()
print(f"Colunas Num√©ricas: \n{colunas_numericas}")
print("\nTotal de colunas num√©ricas: ", len(colunas_numericas))
```

```python
colunas_categoricas = df.select_dtypes(include = ['object']).columns.tolist()
print(f"Colunas Categ√≥ricas: \n{colunas_categoricas}")
print("\nTotal de colunas categ√≥ricas: ", len(colunas_categoricas)) #OBS "Area?"
```

```python
total_nan = df.isna().sum().sum()
print("Total de valores nulos: ", total_nan)

print("\n\nColunas com valores nulos: \n")
for i in df.columns:
    if df[i].isna().sum() > 0:
        print("Coluna: ", i, " - Total de valores nulos: ", df[i].isna().sum())

print("\n\nColunas sem valores nulos: \n")
for i in df.columns:
    if df[i].isna().sum() == 0:
        print("Coluna: ", i)
```

### *Boxplots*

+ First we will plot boxplots for the numeric columns.

#### Comparison of the influence of economic sectors (Gross Added Value).
