# **Importing the Dataset and opening remarks**

## Importing the Dataset and Libraries: üìñ

```python
import pandas as pd
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/IgoSilvaUFPB/projeto_ICD/main/BRAZIL_CITIES.csv'
df = pd.read_csv(url, sep = ';')
df
```

```python
sns.set_palette('Set2')
```

```python
df2 = df.groupby('STATE').sum().sort_values("IBGE_RES_POP")
df2
```

## Line graph showing populations by states.

```python
sns.lineplot(data=df2, x="STATE", y="IBGE_RES_POP")
```

## Scatter plots showing a positive correlation between income and longevity, and income and education.

```python
sns.scatterplot(data=df, x="IDHM_Renda", y="IDHM_Longevidade")
```

```python
sns.scatterplot(data=df, x="IDHM_Renda", y="IDHM_Educacao")
```

## Histograms showing the predominance of different HDI (Human Development Index) levels.

```python
sns.displot(df, x="IDHM", bins=10)
```

```python
sns.displot(df, x="IDHM_Renda", bins=10)
```

```python
sns.displot(df, x="IDHM_Longevidade", bins=10)
```

```python
sns.displot(df, x="IDHM_Educacao", bins=10)
```

## Heatmap and correlation between the different HDI (Human Development Index) levels.

```python
df3 = df2[["IDHM", "IDHM_Renda", "IDHM_Longevidade", "IDHM_Educacao"]]
```

```python
sns.heatmap(df3.corr(), annot=True)
```
